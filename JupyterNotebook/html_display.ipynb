# Drug combination effect

## Grouping by Drug_comb_effect

```python
p_o = pd.read_csv('./Data/pair_pathway_drug_effect.csv')
(p_o.sort_values('gene_sort', ascending=False).set_index(['Drug_comb_effect', 
                                                          'First_Pathway', 'Second_Pathway',])).drop(['gene_sort','group_sort'], axis=1)
```

## Remove repeated gene combination entries without showing pathway 

```python
df = (p_o[['Gene_inhibition', 'P_O', 'Drug_comb_effect', 'Pathway_count']].drop_duplicates().
 sort_values(['Drug_comb_effect','P_O']).set_index('Drug_comb_effect'))

with pd.ExcelWriter('./Data/Drug_effect.xlsx') as w:
    for comb_effect in df.index.unique():
        print(comb_effect)
        df.query("Drug_comb_effect == @comb_effect").to_excel(w, sheet_name=f'{comb_effect}')
df
```

## Grouping by Pair_wise pathway

```python
(p_o.set_index(['First_Pathway', 'Second_Pathway','Drug_comb_effect']).sort_values(['group_sort', 'gene_sort'],
                                                                                  ascending=False).drop(['gene_sort','group_sort'], axis=1))
```
